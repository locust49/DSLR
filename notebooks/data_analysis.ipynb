# DataScience X Logistic Regression

```python
import pandas as pd
import numpy as np
```

## Read dataset from 'datasets/dataset_test.csv' file :

```python
data = pd.read_csv('../datasets/dataset_test.csv', index_col=0)
data.info()
```

## Make the statistics methods

```python
data.describe()
```

```python
# Get numerical data from the dataframe only
data_numeric = data.select_dtypes(include=[np.number])
data_numeric.info()
```

```python
data_numeric
```

```python
# Create the describe dataframe, with the same columns as data_numeric
describe = pd.DataFrame(columns= data_numeric.columns)
```

### count

```python
############################ COUNT ############################
number_of_columns = len(describe.columns)
describe.loc['count'] = [0] * number_of_columns
list_sum = [0] * number_of_columns

for observation in data_numeric.iterrows():
    for i in range(number_of_columns):
        if observation[1][i] is not None and not np.isnan(observation[1][i]):
            describe[describe.columns[i]] += 1
            list_sum[i] += observation[1][i]
        else:
            list_sum[i] += 0
# 
describe
```

### mean

```python
############################ MEAN ############################
describe.loc['mean'] = [0] * number_of_columns
#describe.loc['std'] = [0] * number_of_columns

for i in range(number_of_columns):
    if describe.loc['count'][i] == 0 :
        describe.loc['mean'][i] = np.nan
    else:
        describe.loc['mean'][i] = list_sum[i] / describe.loc['count'][i]
describe
```

### standard deviation

```python
############################ STD ############################
```

### min

```python
describe.loc['min'] = data_numeric.iloc[0]

for observation in data_numeric.iterrows():
    for i in range(number_of_columns):
        if observation[1][i] <= describe.loc['min'][i]:
            describe.loc['min'][i] = observation[1][i]

describe
```

### max

```python
describe.loc['max'] = data_numeric.iloc[0]

for observation in data_numeric.iterrows():
    for i in range(number_of_columns):
        if observation[1][i] >= describe.loc['max'][i]:
            describe.loc['max'][i] = observation[1][i]

describe
```

### 25%

```python
describe.loc['25%'] = [0] * number_of_columns


for i in range(number_of_columns):
    sorted_df = data_numeric.sort_values(by=[describe.columns[i]], ascending=True)
    if describe.loc['count'][i] * 25 / 100 % 2 != 0:
        describe.loc['25%'][i] = sorted_df.iloc[int(describe.loc['count'][i] * 25 / 100)][i]
    else:
        iteration = int(describe.loc['count'][i] * 25 / 100)
        describe.loc['25%'][i] = (sorted_df.iloc[iteration][i] + sorted_df.iloc[iteration + 1][i]) /2

describe
```
